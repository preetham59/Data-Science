```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
diabetes= datasets.load_diabetes()
#dataset is in array format
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
ridge=Ridge()
```

```python
#defining hyperparameters
alphas = np.array([1,0.1,0.01,0.001,0.0001,0])
```

```python
#using gridsearchcv
grid = GridSearchCV(estimator=ridge, param_grid=dict(alpha=alphas))
#fitting the data
grid.fit(diabetes.data, diabetes.target)
```

```python
print(grid.best_score_)
```

```python
print(grid.best_params_)
```

```python
### cross validation and GrisdsearchCV on red wine dataset ###
```

```python
import pandas as pd
```

```python
wine=pd.read_csv('winequality-red.csv')
```

```python
wine.head()
```

```python
wine.isnull().sum()
```

```python
X=wine.iloc[:,0:-1]
y=wine.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
#since we will use CV ,no need to define test size
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
feature_scaler = StandardScaler()
X_train = feature_scaler.fit_transform(X_train)
X_test = feature_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=300, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(estimator=rf, X=X_train, y=y_train, cv=5,scoring='accuracy')
```

```python
print(scores)
print(scores.mean())
print(scores.std())
```

```python
#Gridsearchcv
#defining hyperparameters
grid_param = {
    'n_estimators': [100, 300, 500, 800, 1000],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}
```

```python
grid_2 = GridSearchCV(estimator=rf,
                     param_grid=grid_param,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

```python
grid_2.fit(X_train,y_train)
```

```python
print(grid_2.best_params_)
print(grid_2.best_score_)
```

```python
data = pd.read_csv("ionosphere_csv.csv")
```

```python
data
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
X=data.iloc[:,:-1].values
```

```python
X.shape
```

```python
y=data.iloc[:,-1].values
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(recall_score(y_test,y_pred))
```

```python
print(precision_score(y_test,y_pred))
```

```python
print(f1_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
#GridsearchCV
grid_values = {'penalty': ['l1', 'l2'],'C':[0.001,.009,0.01,.09,1,5,10,25]}
grid_3 = GridSearchCV(lr, param_grid = grid_values,scoring = 'recall')
grid_3.fit(X_train, y_train)
```

```python
print(grid_3.best_params_)
print(grid_3.best_score_)
```

```python
y_grdpred = grid_3.predict(X_test)
```

```python
print(accuracy_score(y_test,y_grdpred))
```

```python
print(precision_score(y_test,y_grdpred))
```

```python
print(f1_score(y_test,y_grdpred))
```

```python
print(confusion_matrix(y_test,y_grdpred))
```
