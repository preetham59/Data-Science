```python
from sklearn import datasets
dataset = datasets.fetch_california_housing()
data = dataset.data
```

```python
import numpy as np
import pandas as pd
headers = dataset.feature_names   #column names
df = pd.DataFrame(data, columns=headers)  
df
```

```python
#defining x and y
y = dataset.target
X = df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.25)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(X_train,y_train)
```

```python
y_pred=lin.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
plot = sns.scatterplot(y_test, y_pred)
plot.set(xlabel='Actual', ylabel='Prediction',label='pred vs actl')
x_plot = np.linspace(0,5,100)
y_plot = x_plot
plt.plot(x_plot, y_plot, color='r')
```

```python
from sklearn.metrics import mean_squared_error
score=-1*mean_squared_error(y_test,y_pred)
print('score_nocv='+str(score))
```

```python
# cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(lin, X, y, cv=10, scoring='neg_mean_squared_error')
print('score_cv =' + str(scores.mean()))
```

```python
loo_scores = cross_val_score(lin, X, y, cv=len(X), scoring='neg_mean_squared_error')
print('score_loocv =' + str(loo_scores.mean()))
```
