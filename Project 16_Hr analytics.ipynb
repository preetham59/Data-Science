```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
le=LabelEncoder()
```

```python
le_count = 0
for col in data.columns[1:]:
    if data[col].dtype == 'object':
        if len(list(data[col].unique())) <= 2:
            le.fit(data[col])
            data[col] = le.transform(data[col])
            le_count += 1
print('{} columns were label encoded.'.format(le_count))
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 5))
data_col = list(data.columns)
data_col.remove('Attrition')
for col in data_col:
    data[col] = data[col].astype(float)
    data[[col]] = scaler.fit_transform(data[[col]])
data['Attrition'] = pd.to_numeric(data['Attrition'], downcast='float')
data.head()
```

```python
data.describe()
```

```python
corr=data.corr()
print(corr)
```

```python
#target of this data is attrition
target = data['Attrition'].copy()
```

```python
#removing target and negatively affecting features
data.drop(['Attrition', 'EmployeeCount', 'EmployeeNumber',
            'StandardHours', 'Over18'], axis=1, inplace=True)
```

```python
data.shape
```

```python
#splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data,
                                                    target,
                                                    test_size=0.25,
                                                    random_state=7)
                                                  
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
lr=LogisticRegression()
```

```python
kf = KFold(n_splits=10, random_state=7)
```

```python
cv_acc_results_lr = cross_val_score(lr, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_lr = cross_val_score(lr, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_lr.mean()*100)
print(cv_auc_results_lr.mean()*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
cv_acc_results_dtc = cross_val_score(dtc, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_dtc = cross_val_score(dtc, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_dtc.mean()*100)
print(cv_auc_results_dtc.mean()*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
cv_acc_results_gnb = cross_val_score(gnb, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_gnb = cross_val_score(gnb, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_gnb.mean()*100)
print(cv_auc_results_gnb.mean()*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
cv_acc_results_knn = cross_val_score(knn, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_knn = cross_val_score(knn, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_knn.mean()*100)
print(cv_auc_results_knn.mean()*100)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
cv_acc_results_svc = cross_val_score(svc, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_svc = cross_val_score(svc, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_svc.mean()*100)
print(cv_auc_results_svc.mean()*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
cv_acc_results_rf = cross_val_score(rf, X_train, y_train, cv=kf, scoring='accuracy')
cv_auc_results_rf = cross_val_score(rf, X_train, y_train, cv=kf, scoring='roc_auc')
print(cv_acc_results_rf.mean()*100)
print(cv_auc_results_rf.mean()*100)
```
