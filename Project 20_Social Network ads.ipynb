```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
sna=pd.read_csv('Social_Network_Ads.csv')
sna
```

```python
sna.columns
```

```python
sna.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
sna.iloc[:,0]=le.fit_transform(sna.iloc[:,0].astype(str))
sna.head()
```

```python
sna.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
sna.iloc[:,1]=le.fit_transform(sna.iloc[:,1].astype(str))
sna.head()
```

```python
x=sna.iloc[:,0:-1]
y=sna.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
#converting to dataframe
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc 1', 'pc 2'])
#combining transformed features and target
ads = pd.concat([principalDf, y], axis = 1)
ads.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kf = KFold(n_splits=20, random_state=0)
results = cross_val_score(lr, x, y, cv=kf, scoring='accuracy')
print(results.mean()*100)
```

```python
results = cross_val_score(lr, x, y, cv=kf, scoring='roc_auc')
print(results.mean()*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
results = cross_val_score(gnb, x, y, cv=kf, scoring='accuracy')
print(results.mean()*100)
```

```python
results = cross_val_score(gnb, x, y, cv=kf, scoring='roc_auc')
print(results.mean()*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
results = cross_val_score(dtc, x, y, cv=kf, scoring='accuracy')
print(results.mean()*100)
```

```python
results = cross_val_score(dtc, x, y, cv=kf, scoring='roc_auc')
print(results.mean()*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
results = cross_val_score(knn, x, y, cv=kf, scoring='accuracy')
print(results.mean()*100)
```

```python

results = cross_val_score(knn, x, y, cv=kf, scoring='roc_auc')
print(results.mean()*100)
```
